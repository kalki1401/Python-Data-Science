```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('Canada.xlsx',sheet_name=1,skiprows=20,skipfooter=2)
df.head(2)
```

- rename column
- drop column
- add column

```python
#rename
cols_to_rename={
    'OdName':'Country',
    'AreaName':'Continent',
    'RegName':'Region',
    'DevName':'Status',
}
df=df.rename(columns=cols_to_rename)
df.head(2)
```

```python
#check no. of nunique values in data
print('Type Unique values',df.Type.nunique())
print('Coverage unique Value',df.Coverage.nunique())
#count unique values in colmn
print('Coverage unique Value\n',df.Coverage.value_counts())
```

```python
cols_to_drop=['AREA','REG','DEV',"Type",'Coverage']
df=df.drop(columns=cols_to_drop)
df.head(2)
```

```python
df
```

```python
df.columns=[str(name).lower() for name in df.columns.tolist()]
df.head()
```

```python
years=list(map(str,range(1980,2014)))
print(years)
```

```python
df[years].sum(axis=1) #axis=1 for calculation on sum
```

```python
df['total']=df[years].sum(axis=1)
df.head()
```

```python
# Visualization

df.isnull().sum()
```

```python
df.isnull().sum().plot(figsize=(10,1),kind='bar')
```

```python
#total immigration visualization
df.total.plot(rot=95,figsize=(25,5),
              xticks=range(0,df.shape[0],2),
title='total immigration')
```

```python
#top 25 countries highest immigration
df.sort_values('total',ascending=False).head(25).plot(kind='bar',y='total',figsize=(15,5),title="top 25 countries highest immigration",color='red',alpha=0.5,xlabel='Country',ylabel='Immigration')
```
