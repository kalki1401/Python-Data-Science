```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax=plt.subplots(nrows=1,ncols=2,figsize=(15,5)) 
t=np.arange(-10,10,.1)
sine=np.sin(t)
cos=np.cos(t)
print(t)
print(sine)
print(cos)
# print(ax[0]) #ax==numpy
ax[0].plot(t,sine,label='sine',color='red',linestyle='',marker='o')
ax[1].plot(t,cos,label='cosine',color='blue')
plt.show()
plt.savefig('sine_cosine.png',dpi=200)
```

plt.plot--> line graphs
plt.bar-->bar graph
plt.pie-->pie chart
plt.scatter-->scatter
